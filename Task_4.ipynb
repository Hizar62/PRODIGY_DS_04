```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
import re
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/drive/MyDrive/Task #4/Sentiment (1).csv')
```

```python
data.head()
```

```python
tweets = data[['sentiment', 'text']]
```

```python
tweets
```

```python
tweets = tweets[tweets['sentiment']!= 'Neutral']
```

```python
tweets.info()
```

```python
#Data Prepocessing
```

```python
train,test = train_test_split(tweets,test_size=0.1)
```

```python
for val in train['text']:
    print(val)
```

```python
pattern = "(@\w+)|(RT\s@\w+:)|(http.*)|(#\w+)"
```

```python
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
def cleaning(data):
    cleaned_tweets = []
    sentiment = []
    for index, row in data.iterrows():
            sentence = re.sub(pattern, '', row.text)
            words = [e.lower() for e in sentence.split()]
            words = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')]
            cleaned_tweets.append(' '.join(words))
            sentiment.append(row.sentiment)

    return cleaned_tweets, sentiment
```

```python
train_tweets, train_sentiment = cleaning(train)
```

```python
train_tweets
```

```python
final_data = {'tweets': train_tweets, 'sentiment': train_sentiment}
```

```python
processed_data = pd.DataFrame(final_data)
```

```python
processed_data
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
processed_data['sentiment'] = labelencoder.fit_transform(processed_data['sentiment'])
```

```python
processed_data
```

```python
from collections import Counter
value_counts = Counter(processed_data['sentiment'])

# Iterate through the results to print each value and its count
for value, count in value_counts.items():
    print(f"'{value}': {count} times")
```

```python
# Converting words into Vectors
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(1,3))
cv.fit(processed_data['tweets'])
```

```python
X_train = cv.transform(processed_data['tweets'])
```

```python
print(X_train.shape)
```

```python
target = processed_data['sentiment'].values
```

```python
target
```

```python
#Cleaning Test Data and Converting Test data into vactors
```

```python
#Cleaning for Test
test_tweets, test_sentiment = cleaning(test)
```

```python
Test_data = {'tweets': test_tweets, 'sentiment': test_sentiment}
```

```python
processed_test_data = pd.DataFrame(Test_data)
```

```python
processed_test_data
```

```python
X_test = cv.transform(processed_test_data['tweets'])
```

```python
print(X_test.shape)
```

```python
X_test.toarray()
```

```python
#Sentiment Analysis model
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
```

```python
classifier.fit(X_train.toarray(),target)
```

```python
y_predict = classifier.predict(X_test.toarray())
```
